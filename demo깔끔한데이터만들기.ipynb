```python
#날씨데이터 
weather = pd.read_csv('c:\\work\\weather.csv')
weather.iloc[:5, :]
```

```python
#깔끔한 데이터
weather_melt = pd.melt(weather, id_vars=['id','year','month','element'],
    var_name='day', value_name='temp')

weather_melt.loc[20:40, ]
```

```python
#pivot_table()메서드로 다시 행을 열로 변환 
weather_tidy = weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns = 'element',
    values='temp'
)

weather_tidy
```

```python
#인덱스를 리셋
weather_tidy_flat = weather_tidy.reset_index() 
#정렬을 지정
weather2 = weather_tidy_flat.sort_values(['year','month','day'], ascending=True) 
```

```python
weather2.info()
```

```python
#형식 변환 
weather_tidy_flat['day2'] = weather_tidy_flat['day'].str.replace('d', '').astype(int)
```

```python
weather_tidy_flat.sort_values(['year','month','day2'], ascending=True) 
```

```python
weather2 = weather_tidy_flat.drop('day', axis=1)
weather2
```
