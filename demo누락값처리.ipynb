```python
#에볼라 발병에 대한 데이터셋
ebola = pd.read_csv('c:\\work\\country_timeseries.csv')
ebola.shape
```

```python
#값이 있는 갯수
ebola.count() 
```

```python
#누락값이 있는 경우 
num_rows = ebola.shape[0]
num_missing = num_rows - ebola.count() 
num_missing
```

```python
#누락값 구하기
import numpy as np 

np.count_nonzero( ebola.isnull() )
```

```python
np.count_nonzero(ebola['Cases_Guinea'].isnull())
```

```python
#value_counts()메서드는 지정한 열의 값의 빈도수 
ebola.Cases_Guinea.value_counts(dropna=False)
```

```python
ebola.Cases_Guinea.value_counts(dropna=True).head()
```

```python
#누락값을 0으로 채우기(10행, 5열) 
ebola.fillna(0).iloc[0:10, 0:5]
```

```python
ebola.iloc[0:10, 0:5]
```

```python
#앞에 있는 값으로 뒤를 채우기
ebola.fillna(method='ffill').iloc[0:10, 0:5]
```

```python
#뒤에 있는 값으로 앞을 채우기
ebola.fillna(method='bfill').iloc[0:10, 0:5]
```

```python
#중간값으로 채우기
ebola.interpolate().iloc[0:10, 0:5]
```

```python
#누락값이 있으면 삭제 
ebola.shape
```

```python
ebola_dropna = ebola.dropna() 
ebola_dropna 
```

```python
#누락값이 포함된 컬럼을 더해서 추가컬럼 생성
ebola['Cases_Multiple'] = ebola['Cases_Guinea'] + ebola['Cases_Liberia'] \
+ ebola['Cases_SierraLeone']
```

```python
ebola_subset = ebola.loc[:, ['Cases_Guinea','Cases_Liberia','Cases_SierraLeone','Cases_Multiple']]
ebola_subset.head(10)
```

```python
#누락값을 제외 
ebola.Cases_Guinea.sum(skipna=True)
```

```python
ebola.Cases_Guinea.sum(skipna=False)
```

```python
#누락값을 제외하고 합계
ebola['Cases_Multiple'] = ebola['Cases_Guinea'].sum(skipna=True) + ebola['Cases_Liberia'].sum(skipna=True)  \
+ ebola['Cases_SierraLeone'].sum(skipna=True) 
```

```python
ebola_subset = ebola.loc[:, ['Cases_Guinea','Cases_Liberia','Cases_SierraLeone','Cases_Multiple']]
ebola_subset.head(10)
```
